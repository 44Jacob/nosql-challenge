### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

````python
# Create a Beautiful Soup object
# Get the HTML content of the visited page
html = browser.html

# Create a BeautifulSoup object and parse the HTML
soup_object = soup(html, 'html.parser')
````

````python
# Find all paragraph elements
paragraphs = soup_object.find_all('p')

# Extract and print the text from each paragraph
for paragraph in paragraphs:
    print(paragraph.get_text())
````

````python
# Iterate over each element in the BeautifulSoup object
for element in soup_object.recursiveChildGenerator():
    if hasattr(element, "text"):
        text = element.text.strip()
        if text:  # Ensure the string is not empty
            print(text)
````

### Step 3: Store the Results

Extract the titles and preview text of the news articles that you scraped. Store the scraping results in Python data structures as follows:

* Store each title-and-preview pair in a Python dictionary. And, give each dictionary two keys: `title` and `preview`. An example is the following:

  ```python
  {'title': "NASA's MAVEN Observes Martian Light Show Caused by Major Solar Storm", 
   'preview': "For the first time in its eight years orbiting Mars, NASA’s MAVEN mission witnessed two different types of ultraviolet aurorae simultaneously, the result of solar storms that began on Aug. 27."
  }
  ```

* Store all the dictionaries in a Python list.

* Print the list in your notebook.

````python
from bs4 import BeautifulSoup
import requests

# Correctly loading HTML content into BeautifulSoup
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'  # Make sure this URL is correct and reachable
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')  # Ensure this object is named correctly and used consistently

articles_list = []

# Correctly loop through each 'h2' tag to find article titles
for title_tag in soup.find_all('h2'):
    article_dict = {'title': '', 'preview': ''}

    article_dict['title'] = title_tag.get_text(strip=True)
    
    preview_tag = title_tag.find_next_sibling('p')
    if preview_tag:
        article_dict['preview'] = preview_tag.get_text(strip=True)
    
    articles_list.append(article_dict)

# Now printing the list should work without error
for article in articles_list:
    print(f"Title: {article['title']}\nPreview: {article['preview']}\n")
````

````python
# Print each article dictionary on a new line for better readability
# Iterate through the list and print each dictionary on a new line
for article in articles_list:
   print(article)
````
